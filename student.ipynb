The primary key is `movie_id` ,so we use it to join the two tables.

```python
#Joining the two tables movie_basics and movie_ratings

pd.read_sql('''
SELECT *,

movie_ratings.movie_id

FROM movie_basics

JOIN movie_ratings

    USING(movie_id)
    
WHERE genres is not null
    
GROUP BY genres

ORDER BY averagerating DESC

;

''',conn)

```

```python
pd.read_sql('''

SELECT *,
genres as genre_null

FROM movie_basics

WHERE genres is null 

;

''',conn)
```
